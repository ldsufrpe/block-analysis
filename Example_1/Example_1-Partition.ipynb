```python
u=var('u');
v=var('v');
t=var('t');
x=var('x');
y=var('y')
```

```python
a = 1/4;
b = 1/7;
t_0 = 7/10;
ϵ = 1/5;
```

```python
P(u,v) = (v^2 - b^2*u^2)^3*((u-t_0)^2+(v-a-ϵ)^2)^3;
```

```python
show(P(u,v).expand())
```

```python
g2(t) = (t-7/10)^2+1/4;
```

### Box $B_{[0, \frac{1}{4}]} =[0, \frac{1}{4}]\times[g_2(1/4), g_2(0)] $

```python
plot(g2, (t, 0, 1/4))
```

```python
(g2(1/4), g2(0))
```

```python
Q1xy =P(u,v).subs(u = (0*x+1/2)/(x+1),v = ((g2(1/4))*y+(g2(0)))/(y+1));
P1xy = Q1xy.numerator();
```

```python
show(P1xy)
```

### Box $B_{[\frac{1}{4}, \frac{1}{2}]} =[\frac{1}{4}, \frac{1}{2}]\times[g_2(1/2), g_2(1/4)] $

```python
plot(g2, (t, 1/4, 1/2))
```

```python
Q2xy =P(u,v).subs(u = ((1/4)*x+(1/2))/(x+1), v = ((g2(1/2))*y+(g2(1/4)))/(y+1));
P2xy = Q2xy.numerator();
```

```python
show(P2xy)
```
