```python
u,v=var('u v');
x,y=var('x y')
t=var('t')
f=-2*sqrt(3)*t - 9*sqrt(2)*t - 18*t + 72*(t^2 + 6*t + 1)/(3*t^2 + 2*t + 3)^(3/2) - 72*(t^2 - 6*t + 1)/(3*t^2 - 2*t + 3)^(3/2) + 8*sqrt(3)/(t + 1) + 8*sqrt(3)/(t - 1) - (2*sqrt(3) + 9*sqrt(2) + 18)/t^2
```

```python
show(f)
```

```python
f_x_y=f.subs({3*t^2 + 2*t + 3:x^2,3*t^2 - 2*t + 3:y^2}).subs(t=(x^2-y^2)/4);
show(f_x_y)
```

```python
assume(x>0);
assume(y>0);
```

```python
f_x_y = f_x_y.simplify(); show(f_x_y.numerator())
```
