```python
load("util.sage")
B6 = load('blocos/bloco_6_linear_cubo.sobj'); B6.dimensions()
```

```python
submatrizes_2x2_dets = B6.minors(2)
p6_t = submatrizes_2x2_dets[-2].expand().combine().simplify()
```

```python
show(p6_t)
```

```python
u, v = var('u, v')
var('t')
assume(u>0)
assume(v>0)

poly_u_v = p6_t.subs({3*t^2 + 2*t + 3:u^2,3*t^2 - 2*t + 3:v^2}).subs(t=(u^2-v^2)/4).factor().numerator()
show(poly_u_v)
```

```python
#Resolve 3*t^2 + 2*t + 3=u^2 em relação a u
sol_u = solve(u**2 == 3*t^2 + 2*t + 3, u)[0].rhs() 

#Resolve 3*t^2 - 2*t + 3=v^2 em relação a v
sol_v = solve(v**2 == 3*t^2 - 2*t + 3, v)[0].rhs()

# obtem os valores de u para t=0 e t=1
u0 = sol_u(t=0)
u1 = sol_u(t=1)

# obtem os valores de u para t=0 e t=1
v0 = sol_v(t=0)
v1 = sol_v(t=1)

u0, u1, v0, v1
```

```python
new_poly_u_v = subs_mobius_2d_uv(poly_u_v, u0, u1, v0, v1)
```
