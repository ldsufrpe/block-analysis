## Analyzing the block B Truncated Tetrahedra

```python
load("../util.sage")
```

```python
blocks = load("blocks")
```

```python
B = blocks.matrix_from_rows_and_columns([1],[1]); 
p1 = B.det().simplify_real().expand().combine()
```

```python
show(p1)
```

```python
p1_simplify = simplify_fractional_radicals(p1); show(p1_simplify)
```

```python
p1_uvt_numerator = p1_simplify.subs({4*t^2 - 4*t + 2:u^2, 6*t^2 - 6*t + 2:v^2}).numerator(); show(p1_uvt_numerator)
```

```python
p1_uvt_numerator = p1_uvt_numerator.canonicalize_radical().expand().combine(); show(p1_uvt_numerator)
```

```python
# Calculate the values of u for t = 0 and t = 1

# Solve 3*t^2 + 2*t + 3 = u^2 for u}
sol_u = solve(u**2 == 4*t^2 - 4*t + 2, u)[0].rhs()

# Solve 3*t^2 - 2*t + 3 = v^2 for v
sol_v = solve(v**2 == 6*t^2 - 6*t + 2, v)[0].rhs()

# Get the values of u for t=0 and t=1
u0 = sol_u(t=0)
u1 = sol_u(t=1/2)

# Get the values of v for t=1/3 and t=1
v0 = sol_v(t=0)
v1 = sol_v(t=1/2)

u0, u1, v0, v1
```

```python
variable_ranges = [(0, 1/2), (u0, u1), (v0, v1)]
```

##### Calculates the Möbius coefficient

```python
coefficient_dict = polynomial_to_dict(p1_uvt_numerator)
```

```python
variables = p1_uvt_numerator.variables()
total_degree  =  [p1_uvt_numerator.degree(j) for j in variables]
n_uplas = [
    [i, j, k]
    for i in range(total_degree[0] + 1)
    for j in range(total_degree[1] + 1)
    for k in range(total_degree[2] + 1)
]
```

```python
print(variables)
```

```python
import time


# Capture o tempo de início
start_time = time.time()
sign_coeff_set = set()

for i, degree_per_variable in enumerate(n_uplas):
    
    c = moebius_coefficient(p1_uvt_numerator, degree_per_variable, total_degree, variable_ranges, coefficient_dict)

    if i % 100 == 0:
        elapsed_time = time.time() - start_time
        elapsed_time_minutes = elapsed_time / 60
        print(f"Iteração: {i}, Tempo decorrido: {elapsed_time_minutes:.2f} minutos")
        print("Conjunto parcial de coeficientes:", sign_coeff_set)  
        print("\n")
    
    
    if c.is_zero():         
         continue

    else:
        sign_coeff = sign(c.simplify_full())
        sign_coeff_set.add(sign_coeff)

    if 1 in sign_coeff_set and -1 in sign_coeff_set:
        
        print(f"Encerrando na iteração {i} pois o conjunto contém 1 e -1.")
        break
        
    
print("Conjunto final de coeficientes:", sign_coeff_set)  
```
