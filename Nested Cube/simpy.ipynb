```python
C12_diff = load('C12_diff')
```

```python
C12_diff_sp = C12_diff._sympy_()
```

```python
import sympy as sp
x_v, x_u, x_t = C12_diff_sp.free_symbols
a, b, c, d, e, f = sp.symbols('a, b, c, d, e , f')
```

```python
#(x_v + 1), (x_t + 1)
C12_diff_sp_new = C12_diff_sp.subs(x_v + 1, a).subs(x_u + 1, b).subs(x_t + 1, c).subs(sqrt(6)*x_v/3 + 1, d).subs((sqrt(3)*x_u + 2*sqrt(2)), e)
```

```python
C12_diff_sp_new_nu = sp.numer(sp.simplify(C12_diff_sp_new))
```

```python
C12_diff_sp_new_nu
```

```python
poly_C12_diff = C12_diff_sp_new_nu.subs(a, x_v + 1).subs(b, x_u + 1).subs(c, x_t + 1).subs(d, sqrt(6)*x_v/3 + 1,).subs(e, (sqrt(3)*x_u + 2*sqrt(2)))
```
